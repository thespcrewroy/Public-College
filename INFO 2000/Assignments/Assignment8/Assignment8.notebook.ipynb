```python
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Plotting

```python
'''
Create a simple line plot using Matplotlib 
* function y = 4x^2 + 2
* range = [-20,20]
* Title
* Label the x and y axis
* Add appropriate xticks
'''

x = np.linspace(-20, 20) # create an array of x values
fig,axis = plt.subplots() # create a figure and axis
axis.plot(x, 4*x**2 + 2, "b--") # plot the function
fig.suptitle(r"$y = 4x^2 + 2$") # add a title
axis.set(xlim=[-20, 20], 
         xlabel="x-axis",
         ylabel="y-axis",
         xticks=x) # set x-axis limits, label and ticks
plt.xticks(np.arange(-20, 21, 5)) # set x-axis ticks at intervals of 5
plt.show() # show the plot
```

# Scatter Plot

```python
'''
Generate a scatter plot
* 100 random points
* xrange = [0,100]
* yrange = [0,100]
* markercolor = blue
* markershape = circles
* marketedges = red
* create a legend
'''

plotx = np.random.randint(0, 100, 100) # generate 100 random x values
ploty = np.random.randint(0, 100, 100) # generate 100 random y values
fig,axis = plt.subplots() # create a figure and axis
axis.scatter(plotx, 
             ploty, 
             color="blue", 
             marker="o", 
             edgecolors="red",
             label="Blue") # plot the scatter plot
axis.set(xlim=[0, 100], ylim=[0, 100]) # set x and y range to [0,100]
axis.legend() # add a legend
plt.show() # show the plot
```

# Bar Chart

```python
'''
Create a bar chart to compare the total points per game of the following teams
* LSU: 45.29 tp/g
* Ole Miss: 41.67 tp/g
* Georgia: 40.14 tp/g
* Missouri: 33.86 tp/g
* Tennessee: 33.50 tp/g

Style appropriately:
* Customize the bars
* Customize the lebels
* Customize the title
'''
fig, ax = plt.subplots() # create a figure and axis
x_labels = np.array(["LSU", "Ole Miss", "Georgia", "Missouri", "Tennessee"]) # x-axis labels
x_locs = np.array([0, 1, 2, 3, 4]) # x-axis locations
y_values = np.array([45.29, 41.67, 40.14, 33.86, 33.50]) # y-axis values

bar_width = 0.3 # width of the bars
bar_sep = 0.1 # separation between bars

ax.bar(x_locs - bar_width/2 - bar_sep,
       y_values,
       width=bar_width,
       facecolor="r",
       label="Red") # plot the bars
ax.set(xticks=x_locs - bar_width/2 - bar_sep,
       xticklabels=x_labels,
       xlabel="Teams",
       ylabel="Total Points/Game",
       title="Total Points Per Game Comapred to Teams") # set x-axis labels, y-axis label, and title
ax.legend() # add a legend
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.show()
```

# Subplots

```python
fig, axs = plt.subplots(2, 2) # create a 2x2 subplot
x = np.linspace(-20, 20) # create an array of x values
a = [1, 2, 3, 4] # values of a
c = [1, 2, 3, 4] # values of c

for i in range(2):
    for j in range(2):
        y = a[i] * x**2 + c[j] # calculate y values for each subplot
        axs[i, j].plot(x, y) # plot the line in the corresponding subplot
        axs[i, j].set_title(f"a={a[i]}, c={c[j]}") # set the title for each subplot

plt.show() # show the plot
```
