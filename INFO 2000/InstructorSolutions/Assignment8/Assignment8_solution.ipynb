```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1 Basic Plotting
Create a simple line plot using Matplotlib that represents the function
y = 4x2 + 2
in the range [-20,20]. Add a title, label the x and y axis, and add appropriate x-ticks.

```python
# 1: plotting y = 4x^2 + 2
x = np.linspace(-20,20,50)
y = 4 * x**2 + 2
fig, ax = plt.subplots(figsize=(6, 4), tight_layout=True) #plt.plot(y,x)
ax.plot(x,y,c='green',marker='o',mfc ='black',mec = 'y',label='Sales',linewidth=5) # c = plot color, mfc = marker face color, mec = marker edge color, etc
ax.set_title('A line plot')
ax.set_xlabel('X-values')
ax.set_ylabel('Y-values')
# set the x-ticks every half decade
ticks = np.arange(-20,25,5)
ax.set(xticks = ticks)
ax.legend(loc = 1)
plt.tight_layout()
plt.show()
```

# 2 Scatter Plot
Generate a scatter plot with 100 random points in the range [0,100] for both x and y axes. Customize the

# 4 Subplots
The generic form of the equation in the basic plotting question is y = ax^2 + c
Use four different values of a and/or c to generate 4 values for y, and plot them in one figure but in different
subplots subplots. In other words, you will have a 2 by 2 plot with each of the lines in its own subplot. Link to
documentation and examples: plotting documentation. Customize the plots with appropriate labels, titles, and
colors.

```python
#4 four plots
import random
# generate the subplot figure and axes
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(10, 10),
                        layout="constrained") # constrained layout prevents overlap of labels and prevents overflow outside figure

# pick 4 values of a and c randomly
a_list = random.sample(list(np.arange(10)),4)
c_list = random.sample(list(np.arange(0,20,2)),4) # creating a uniform range using numpy, then casting it into a list, then sampling 4 items from the resulting list using the random library

print(f"a: {a_list}, c:{c_list}")
x = np.array(range(50))

# let's use a for loop to generate our values and plot them

for i in range (2):
    for j in range(2):
        # ok, not the most straightforward way to pick numbers from an array/list
        # but we need to convert our linear (1 -D) indexing for a and c, into a 2 D indexing so that it can work with the current for loop
        idx = 2 * i + j 
        a = a_list[idx]
        c = c_list[idx]

        y = a * x**idx + c
        ax = axs[i,j]

        ax.plot(x,y,c='green',marker='o',mfc ='black',mec = 'y',label=f'x^{idx}',linewidth=2) # c = plot color, mfc = marker face color, mec = marker edge color, etc
        ax.set_title(f'a = {a}, c = {c}')
        ax.set_xlabel('X-values')
        ax.set_ylabel('Y-values')

        # set the x-ticks every decade
        ticks = np.arange(0,51,10)

        ax.set(xticks = ticks)
        ax.legend()
        plt.tight_layout()
plt.suptitle("Several line plots")
plt.show()
plt.tight_layout()
        
```
