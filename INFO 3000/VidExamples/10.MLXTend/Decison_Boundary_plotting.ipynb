#### Quick tutorial on "mlxtend" library ####

mlxtend is a library in Python that provides a lot of useful tools for machine learning. One of these is plot_decision_regions, which is a convenient tool for visualizing decision boundaries in up to two dimensions.

Installing:<br>
<em>pip install mlxtend</em>

Reference: https://rasbt.github.io/mlxtend/user_guide/plotting/plot_decision_regions/

The plot_decision_regions function is creating a mesh of points, predicting the class for each point, and using this to color-code the background. This forms the decision boundary. The actual data points are then overlayed on top.

Note: The plot_decision_regions function only works for datasets with two features (since we are visualizing in 2D). If your dataset has more than two features, you will have to reduce the dimension (using techniques such as PCA) before using this function.

Here's a quick tutorial on how you could use this function to visualize a decision boundary with logistic regression. However, this is applicable for most models.

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Import the decision boundary function from mlxtend
from mlxtend.plotting import plot_decision_regions
```

```python
# Generate some data

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, n_samples=100)
```

```python
# Fit a logistic regression model

model = LogisticRegression() # Create a logistic regression model
model.fit(X, y) # Fit the model to the data
```

```python
# Plot the decison boundary
# X,y should be fed to the function as numpy arrays

plot_decision_regions(X, y, clf=model, legend=2) # The function expects features(X), labels(y), model used, as inputs.
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary of Logistic Regression')
plt.show()
```
